```python
classifier = pipeline("sentiment-analysis", model="bert-base-uncased")

text_input_1 = "I really enjoyed this movie. The acting was great."

bert_output = classifier(text_input_1)

print("=== BERT Sentiment Classification ===")
print("Input:", text_input_1)
print("Output:", bert_output)
print("--------------------------------------\n")
```

```python
generator = pipeline("text-generation", model="gpt2")

text_input_2 = "Once upon a time in a quiet village"

gpt2_output = generator(text_input_2, max_length=50, num_return_sequences=1)

print("=== GPT-2 Text Generation ===")
print("Input:", text_input_2)
print("Output:", gpt2_output[0]['generated_text'])
print("--------------------------------------\n")
```

```python
summarizer = pipeline("summarization", model="t5-small")

text_input_3 = """
Machine learning models are trained on large datasets to discover patterns
and make predictions. Transformers represent one of the most powerful architectures
because they can process long sequences and capture global context using self-attention.
"""

t5_output = summarizer(text_input_3, max_length=50, min_length=10, do_sample=False)

print("=== T5 Summarization ===")
print("Input:", text_input_3)
print("Output:", t5_output[0]['summary_text'])
print("--------------------------------------\n")
```
